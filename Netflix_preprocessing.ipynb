# Récupération date de sortie netflix avec l'API IMDb gratuite

```python
import requests
from Keys_APIs import get_API_key_TMDb

API_KEY = get_API_key_TMDb()

def get_release_date_tmdb(title):
    try:
        # Étape 1 : recherche du film par titre
        search_url = "https://api.themoviedb.org/3/search/movie"
        params = {
            "api_key": API_KEY,
            "query": title
        }
        response = requests.get(search_url, params=params).json()
        results = response.get("results")
        if not results:
            return None

        # Prend le premier résultat
        movie_id = results[0]['id']

        # Étape 2 : Récupération des dates de sortie
        release_url = f"https://api.themoviedb.org/3/movie/{movie_id}/release_dates"
        release_response = requests.get(release_url, params={"api_key": API_KEY}).json()
        releases = release_response.get("results", [])

        # Recherche la date pour US ou mondial
        for country_data in releases:
            if country_data['iso_3166_1'] in ['US', 'FR', 'GB']:
                for entry in country_data['release_dates']:
                    return entry['release_date'][:10]  # format 'YYYY-MM-DD'

        return None

    except Exception as e:
        print(f"Erreur pour {title}: {e}")
        return None

print(get_release_date_tmdb("The Irishman"))
```

```python
for idx, row in data[data['Release Date'].isna()].iterrows():
    title = row['Title']
    date = get_release_date_tmdb(title)
    data.at[idx, 'Release Date'] = date
```

```python
data.fillna({
    'Hidden Gem Score': -1,
    'Country Availability': "Unknown",
    'Writer': "Unknown",
    'Actors': "Unknown",
    'View Rating': "Uknown",
    'IMDb Score': data['IMDb Score'].mean(),
    'Rotten Tomatoes Score': data['Rotten Tomatoes Score'].mean(),
    'Metacritic Score': data['Metacritic Score'].mean(),
    'Awards Received': 0,
    'Awards Nominated For': 0,
    'Boxoffice': "$0",
    'Production House': "Unknown",
    'IMDb Votes': data['IMDb Votes'].mean(),
    'Trailer Site': "Unknown"
}, inplace=True)
```

```python
from datetime import datetime

def convert_date_format(date_str):
    try:
        return datetime.strptime(date_str, "%d %b %Y").strftime("%Y-%m-%d")
    except:
        return date_str

data['Release Date'] = data['Release Date'].apply(convert_date_format)
```
